### 4.2 Q2: Analyze how long does it take for a video to become trendy.

In other words, how many days does the video need to become trending?

First of all, let's convert time of the column 'publishedAt' to a simple date format, which is more familiar for us. For example, now that is like 2020-08-21T03:58:10Z and, according to the https://www.w3.org/TR/NOTE-datetime, Z means UTC (Coordinated Universal Time), and it will be 03:58:10 21st of August 2020 by Coordinated Universal Time. The amount of needed time to become trendy we will count by days.

```python
import datetime #we will work with dates, so we need datetime

#creating new column with needed date format 
all_yt["publishing_day"] = all_yt["publishedAt"].apply(
    lambda x: datetime.datetime.strptime(x[:10], "%Y-%m-%d"))

#all_yt.head(10)
```

```python
#adding new column for trending day, where we just change the format 
all_yt["trending_day"] = all_yt["trending_date"].apply(
    lambda x: datetime.datetime.strptime(x[:10], '%y.%d.%m').strftime('%Y-%m-%d'))
```

```python
#since we have new columns, we don't need old. Let me drop them
all_yt.drop(labels='publishedAt', axis=1, inplace=True)
all_yt.drop(labels='trending_date', axis=1, inplace=True)
```

```python
#let's see what we've done
all_yt.head(10)
```

```python
#checking types
all_yt.info()
```

```python
#changing type for 'trending_day' column
all_yt['trending_day'] = pd.to_datetime(all_yt['trending_day'])
```

```python
#creating a new column to see how many days videos need to become popular
all_yt['days_to_trend'] = (all_yt['trending_day'] - all_yt['publishing_day']).dt.days
#by using .dt.days we will get rid of 'days' word
```

```python
all_yt.head(10)
```

Our data is ready to be analyzed! To show how many days video need we'll use boxplot.

```python
#sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,6))
ax = sns.boxplot(x=all_yt["days_to_trend"], #creating boxplot
                 palette="husl") #setting palette

#setting title and label
#plt.xticks(np.arange(0, 25, 1)) #if you want
plt.xlabel("Number of days", size=10)
plt.title("How many days video needs to become trending", size=14)
ax.xaxis.grid(True) #adding grid
```

On average, the video needs 3 days to become trending. Moreover, most of the videos need from 0 to 9 days to appear in trending list. We can also notice some trending videos that became trendy only after 10 or even 21 days after they were published.

 ### 4.3 Q3: Analyze the correlations of trends.

Each video has likes, dislikes, comments, ratings, dates(when it was published and when it became trending). For me, it it pretty interesting if there is any link between them. 

It is believed that bad publicity is still good P.R. But can the bad publicity may bring to you many views and comments?

Let me try to answer it by following:

```python
#First of all, let's correlate each column and see how it looks by the table
all_yt.corr()
```

```python
#Secondly, let's illustrate the above correlation by heatmap

title_map = [x.replace('_', ' ') for x in #replacing '_' by regular space
            list(all_yt
                 .select_dtypes(include=['number', 'bool']) #selecting column, where data types are number or boolean
                 .columns.values)] #getting column names

fig, ax = plt.subplots(figsize=(12,7))
sns.heatmap(all_yt.corr(), annot=True, #annot=True is adding labels to the rectangles
            xticklabels=title_map, yticklabels=title_map,  #setting labels
            cmap=sns.cubehelix_palette(start=2.8, rot=.1, as_cmap=True)) #setting palette

ax.set(title="Correlations of Trends")
#ax.set_xticklabels(title_map, rotation = 45)
```

In this heatmap, the darker color indicates higher positive correlation, while lighter - lower. The highest possible indicator is 1 and the lowest is -1. 

So, by looking at this heatmap we can draw conclusions on the following:

+ Number of likes and views are highly correlated with the number of 0.97, which means that if in the trending video there is many views, there must be many like too.
+ Correlation indicator for number of views and comments, likes and comments has the same high number - 0.93. Thus, if the number of comments of the trending video increases, number of views and likes goes up too.
+ Bad publicity might help your video become trending, but with big amount of dislikes it will not get a high number of comments and views (indicators are 0.16 and 0.24 respectively)
+ The lowest correlation indicator is -0.064 for comment count and the number of days taken for the video to become trending.
